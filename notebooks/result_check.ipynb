```python
from pathlib import Path
```

```python
import pandas as pd
pd.set_option('display.max_rows', 300)
```

```python
OUTPUT_DIR = Path('../outputs')
df = pd.read_csv(OUTPUT_DIR.joinpath('result.csv'))
df.columns = ['code', 'name', 'teacher', 'room', 'period', 'course', 'degree', 'class', 'status', 'dulation']
df.head()
```

```python
# 曜日で並べてソートしてエクセルで保存
df.sort_values(by=['period']).to_excel(OUTPUT_DIR.joinpath('result.xlsx'), index=False)
```

```python
# 教室被りチェック
periods = sorted(df['period'].unique().tolist())
for p in periods[0:1]:
    _df = df[df['period'] == p]
    print(_df['room'].value_counts())
```

```python
courses = sorted(df['course'].unique().tolist())
```

```python
# 重複してしまう講義の例を見る
for p in periods:
    print('-' * 30)
    print(p)
    for c in courses:
        id_dup = df[df['course'] == c].duplicated(subset=['period', 'degree', 'class', 'status']).index
        df_dup = df.loc[id_dup]
        print(c, df_dup[df_dup['period'] == p]['code'].values)
```
