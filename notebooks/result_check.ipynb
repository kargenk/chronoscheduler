```python
from pathlib import Path
```

```python
import pandas as pd
pd.set_option('display.max_rows', 300)
```

```python
OUTPUT_DIR = Path('../outputs/M/first')
df = pd.read_csv(OUTPUT_DIR.joinpath('result.csv'))
df.columns = ['code', 'name', 'teacher', 'room', 'period', 'course', 'degree', 'class', 'status', 'duration']
df['teacher'] = df['teacher'].apply(lambda x: pd.Series(x))
df.head()
```

```python
# 教室被りチェック
periods = sorted(df['period'].unique().tolist())
for p in periods[0:1]:
    _df = df[df['period'] == p]
    print(_df['room'].value_counts())
```

```python
courses = sorted(df['course'].unique().tolist())
```

```python
# 学生目線で重複してしまう講義の例を見る
for p in periods:
    print('-' * 30)
    print(p)
    for c in courses:
        id_dup = df[df['course'] == c].duplicated(subset=['period', 'degree', 'class', 'status']).index
        df_dup = df.loc[id_dup]
        print(c, df_dup[df_dup['period'] == p]['code'].values)
```

```python
# 授業コードと部屋の対応
df.sort_values(by='period')
code_to_room = dict(df[['code', 'room']].values)
code_to_room[1]
```

```python
df_shaped = df.sort_values(by=['period']).set_index(['period', 'course', 'degree', 'class']).sort_index()
df_shaped = pd.get_dummies(df_shaped, columns=['room'], prefix='', prefix_sep='')
df_shaped
```

```python
# 時限、クラスでまとめてソートしてエクセルで保存
df_output = df.sort_values(by=['period']).set_index(['period', 'course', 'degree', 'class']).sort_index()
df_output = df_output.reindex(columns=['code', 'name', 'teacher', 'status', 'duration', 'room'])
df_output.to_excel(OUTPUT_DIR.joinpath('result.xlsx'))
```
